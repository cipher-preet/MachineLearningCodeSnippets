```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

url = "https://www.indiamart.com/?srsltid=AfmBOoqfq5lI73boU9O9pVophUzUhnzHpNdsMcF-nWTbkq2EVTRmScO"

webpage = requests.get(url,headers=headers, timeout=100).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
#soup.find_all('h1')

print(soup.prettify())
```

```python
soup.find_all('p')
```

```python
name = []

for i in soup.find_all('p'):
    name.append(i.text)

print(name)
```
